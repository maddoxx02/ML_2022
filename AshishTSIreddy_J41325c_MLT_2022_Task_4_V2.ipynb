# 1st Round of Processing (i.e. Stemming & Lemmatizing)

```python
def stle(chapter):
    
    lemer = WordNetLemmatizer()
    stemer = PorterStemmer()
    
    [lemer.lemmatize(token) for token in chapter]
    [stemer.stem(token) for token in chapter]
    
    return chapter
```

# 2nd Round of Processing Data (Lower Case, Stop Words, Tokenization) 

```python
def interprocessing(indata):
    
    temptoken = TreebankWordTokenizer().tokenize(indata) #Tokenization

    temptoken = [token.lower() for token in temptoken] # Lower Case
    
    stopwordsc = stopwords.words("English") #Calling StopWords 
    temptoken = [token for token in temptoken if token not in stopwordsc] #Clearing Stopwords 
    
    return temptoken
```
